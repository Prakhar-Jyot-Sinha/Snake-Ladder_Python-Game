# <u>Assignment Sankes & Ladders</u>

**Designing and implementing a Snakes and Ladders game in Python.**

- The game consists of 2 modes: Auto mode and Cheat/Manual mode.

- For auto mode, a player will type `roll` and a random number will be generated
  imitating the roll of dice. 
 
- There are 6 face dice which are being rolled by the player to their
  chance.
  
- For a manual mode, instead of entering roll, players can enter any number between `1 and 20`.
  
- The player starts from 0 and has to reach the final position (in our case,
  it's 100).
  
- There are some ladders which turn out to be lucky for the player as
  they shorten the way.
  
- There are some snakes present in between the game which turns out
  to be the enemy of the player as they just lengthen their way to 100.
  
- The position of snakes and ladders are stored in start:end format
  where start denotes the position of ladder/snake and end denotes the
  value that it reaches.
  
- There are only 2 players in the game and no spectators 

- During any playerâ€™s chance, if he/she enters `quit`, then
  the game will instantaneously end for that particular player and the other
  player will win the game.
  
`Position of snakes: {17 : 7, 54 : 34, 62 : 19, 98 : 79}`
`Position of ladders: {3 : 38, 24 : 33, 42 : 93, 72 : 84}`
<hr>

`Concepts used:`
- *List, Tuple, Dictionary*
- *Random number generation*
- *Iterations, conditional statements and functions*

```python
# importing required extentions
import random
```

```python
# Starting the game with welcoming Players
print("###### Welcome to Snakes & Ladders Game ######")
print("= "*50)

# Asking names of the players
player1=input("Enter the name of Player1: ").upper()
player2=input("Enter the name of Player2: ").upper()
print(" ")

print("###### Let us Start #####")
print(" ")

# Brief description of the game
print("THERE ARE BASICALLY 2 MODES AUTO_MODE & CHEAT_MODE \nFOR AUTO_MODE Enter 'Roll' to roll the DICE\nAnd for CHEAT_MODE Enter any integer between '1 to 20' \nYou can also enter QUIT to leave the game in between.")
print("= "*50)

# defining the list with names and their current position
pos1=0
pos2=0
players = [[player1,pos1],[player2,pos2]]

# creating list for cheat mode
CHEAT_MODE=[str(x) for x in range(1,21)]

# defing positions of snakes and ladders in the dictionary as keys, 
# and values as to which position player will drop or climb.
snakes= {17:7,54:34,62:19,98:79}
ladders= {3:38,24:33,42:93,72:84}

# starting the loop to start the game and initializing `i` with 0 to start with player_1
i=0
while True:
    
    # taking input from the player
    inp = input(f"{players[i][0]} its your turn. Enter roll or an integer between 1 and 20: ").lower()
    print(" ")
    
    # condition for roll as input
    if inp == 'roll':
        x=random.randint(1,6) # rollong the dice
        print("Your dice rolled to ",x)
        
        # condition if current position + x value turns out to be greater than 100
        if (players[i][1] + x) > 100:
            print("OH!!.. you are close to win but did not make it to 100th position, you went ahead so you stay at Same Position \nGOOD LUCK for your next turn")
            print(" ")
            pass
        else:
            players[i][1] += x
        print(f"{players[i][0]} your current possition is {players[i][1]}")
        print(" ")
        
    #condition for QUIT as input
    elif inp == "quit":
        print(f"{players[i][0]} chooses to QUIT the game")
        print(" ")
        print(f"HURRAYYY!!!!!!\n\n{players[((i+1)%2)][0]} WINS the GAME.......")
        print(" ")
        print("= "*50)
        print("****** GAME ENDS ******")
        print("###### THANK YOU FOR PLAYING ######")
        print("# "*50)
        break
    
    # if player wants to enter CHEAT MODE
    else:
        if inp in CHEAT_MODE:
            x = int(inp) #type casting to int
            
            # condition if current position + x value turns out to be greater than 100
            if (players[i][1] + x) > 100:
                print("OH!!.. you are close to win but did not make it to 100th position, you went ahead so you stay at Same Position \nGOOD LUCK for your next turn")
                print(" ")
                pass
            else:
                players[i][1] += x
            print(f"{players[i][0]} your current possition is {players[i][1]}")
            print(" ")
            
        # if player enter invalid input he/she is asked to enter again    
        else:
            print(f"{players[i][0]} Invalind Input please enter again")
            print(" ")
            continue
    
    # if player reaches 100 position he/she wins and game ends
    if players[i][1] == 100:
        print(f"HURRAYYY!!!!!! \n\n{players[i][0]}WON THE GAME....")
        print(" ")
        print("= "*50)
        print("****** GAME-ENDS ******")
        print("###### THANK YOU FOR PLAYING ######")
        print("# "*50)
        break
        
    # condition if player encounters ladder or snake
    if players[i][1] in ladders:
        print("..WoW!! how lucky you got a LADDER..")
        print(" ")
        x=ladders[players[i][1]]
        players[i][1]=x
        print(f"{players[i][0]} you climbed to {players[i][1]}")
        print(" ")
        
    elif players[i][1] in snakes:
        print("..OH SHIT!! You met the SNAKE..")
        print(" ")
        x=snakes[players[i][1]]
        players[i][1]=x
        print(f"{players[i][0]} you dropped to {players[i][1]}")
        print(" ")
    
            
    
    # loop incrementation 
    # rotating turns between the players till game ends
    
    i = (i+1)%2
    
```
